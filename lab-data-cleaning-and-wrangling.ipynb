```python
def standard_scales(df):
    cols = df.columns
    from sklearn.preprocessing import StandardScaler
    transformer = StandardScaler().fit(df)
    x_standardized = transformer.transform(df)
    return pd.DataFrame(x_standardized, columns=cols)
```

```python
import datetime

def get_months(x):
    x = x.strftime("%m")
    return x
```

```python
import datetime

def get_years(x):
    x = x.strftime("%Y")
    return x
```

```python
def get_day(x):
    x = x.strftime("%j")
    return x
```

```python
def get_week(x):
    x = x.strftime("%U")
    return x
```

```python
customer_df['month'] = list(map(get_months, customer_df['effective_to_date']))
```
